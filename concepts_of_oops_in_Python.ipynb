```python
class Calculator:
  # def add(self,a):
  #   return a
  def add(self,a,b=0,c=0): #so you assigned the value for b and c as 0 , so anytime u miss to assign the value for them python by default consider the as 0
    return a+b+c
obj = Calculator()
print(obj.add(5))
print(obj.add(4,5))
print(obj.add(4,5,6))
```

from abc

#This module helps us force some rules on the classes

import ABC

ABC is a class Inside abc module

import abstractMethod

it is used to mark a method as be written inside child


```python
from abc import ABC , abstractmethod
class Vehicle(ABC): #ABC is your abstract base class so u can call it as parent class
#once u inherit ABC in your child class , Vehicle , Vehicle also becomes abstract class
  @abstractmethod #this method must be written in child class
  def start(self):
     pass
v = Vehicle()
v.start()
#This is considerd as an incomplete class by python that is why it cannot create an object for this
#
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass


# Child class implementing abstract method
class Car(Vehicle):

    def start(self):
        print("Car engine started")


# Create object of child class
v = Car()
v.start()
```

Create a class ATM
The class must :
Store pin and balance
Have method check_pin(input_pin)
Have method withdraw(amount)
Withdraw will be done only when amount is sufficient

```python
class ATM:
  def __init__(self,pin,balance):
    self.pin = pin
    self.balance = balance
  def check_pin(self,input_pin):
    return self.pin == input_pin
  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance = self.balance-amount
      print("Withdrawl",amount)
      print("Remaining balance",self.balance)
    else:
      print("Insufficient fund")
a = ATM(1234,5000)
if a.check_pin(1234):
  a.withdraw(2000)
else:
  print("Wrong Pin")
```
